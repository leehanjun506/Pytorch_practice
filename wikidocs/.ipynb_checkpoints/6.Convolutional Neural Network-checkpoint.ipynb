# 합성곱 신경망(Convolutional Neural Network)
---
## 1. 합성곱과 풀링(Convolution and Pooling)
합성곱 신경망은 크게 합성곱층과(Convolution layer)와 풀링층(Pooling layer)으로 구성된다.
### 1. 합성곱 신경망(CNN)의 대두
이미지를 1차원 벡터로 변환하여 다층 퍼셉트론의 입력층으로 사용하면 변환 전에 가지고 있던 공간적인 구조(spatial structure)정보가 유실된다. 따라서 이미지의 공간적인 구조 정보를 보존하면서 학습할 수 있는 합성곱 신경망을 사용한다.
### 2. 채널(Channel)
이미지는 (높이,너비,채널)이라는 3차원 텐서이다. 높이는 이미지의 세로 방향 픽셀 수, 너비는 이미지의 가로 방향 픽셀 수, 채널은 색 성분
### 3. 합성곱 연산(Convolution operation)
합성곱층은 합성곱 연산을 통해서 이미지의 특징을 추출하는 역할을 한다.
아래의 예제는 스트라이드가 2일 경우에 5 * 5 이미지에 합성곱 연산을 수행하는 3 * 3 커널의 움직임을 보여주고 최종적으로 2 * 2 특성 맵을 얻는다.  

### 4. 패딩(Padding)
특성 맵(feature map)의 크기가 입력보다 작아지는 것을 방지하기 위해 패딩(padding)을 사용한다.

### 6. 3차원 텐서의 합성곱 연산
다수의 채널을 가진 입력 데이터를 가지고 합성곱 연산을 하면 커널의 채널 수도 입력의 채널과 같아야한다.  

### 7. 풀링(Pooling)
풀링은 특성 맵을 다운샘플링하여 특성 맵의 크기를 줄여준다.  
이때 풀링 연산으로 최대 풀링(max pooling)과 평균 풀링(average pooling)이 사용된다.
